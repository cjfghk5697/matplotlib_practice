<h1>Matplotlib 라이브러리의 기초</h1>
Matplotlib이란 다양한 데이터를 시각화 할 수 있도록 도와주는 라이브러리이다. 간단한 데이터 분석에서부터 인공지능 모델의 시각화까지 활용도가 매우 높다.

일단 간단한 직선 그래프 그리기부터 하겠다.

```python
import matplotlib.pyplot as plt

x=[1,2,3]
y=[1,2,3]
plt.plot(x,y)
plt.title("My Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

그래프를 savefig를 이용해 저장했다.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0, np.pi*10, 500) #PI*10 너비에, 500개의 점을 균일하게 찍기
fig, axes= plt.subplots(2,1) # 2개의 그래프가 들어가는 Figure 생성
axes[0].plot(x,np.sin(x)) #첫 번째 그래프는 사인(sin) 그래프
axes[1].plot(x,np.cos(x)) # 두 번째 그래프는 코사인(cos) 그래프
fig.savefig("sin&cos.png")
```

subplots를 이용해 2개의 그래프가 들어가게 한다. 

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-9,10)
y=x**2

#라인 스타일로는 '-', ':', '-.', '--' 등이 사용될 수 있습니다.
plt.plot(x,y, linestyle=":", marker="*")
# X축 및 y축에서 특정 범위를 자를 수도 있다.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-9,10)
y1= x**2
y2= -x
plt.plot(x, y1, linestyle="-.", marker="*", color="red", label="y=x*x")
plt.plot(x,y2, linestyle=":", marker="o",color="blue",label="y=-x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(
    shadow=True,
    borderpad=1
)
plt.show()
```

레이블 정보가 표현되는 것을 legend라고 한다. 위에 그래프에 박스에 잘 표현 된 것을 확인 할 수 있다.
